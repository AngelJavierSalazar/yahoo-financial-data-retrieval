Apple Inc

```python
aapl = YahooFinancials('AAPL')
aapl
```

Historical data retrieval: daily, weekly, monthly

```python
hist = aapl.get_historical_price_data('2022-01-01', '2022-06-30', 'daily')
print(hist)
```

```python
hist = aapl.get_historical_price_data('2022-01-01', '2022-06-30', 'weekly')
print(hist)
```

```python
hist = aapl.get_historical_price_data('2022-01-01', '2022-06-30', 'monthly')
print(hist)
```

```python
import pandas as pd
pd.set_option('display.max_columns', 7)

hist_df = pd.DataFrame(hist['AAPL']['prices']).drop('date', axis=1).set_index('formatted_date')
print(hist_df)
```
